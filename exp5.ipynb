```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Box Filter
img = cv.imread("/Users/ved/dip/images/img1-1.bmp", 0)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
filter_size = 5
pad_img = np.pad(img, pad_width=2, mode="constant", constant_values=0)
```

```python
rows, cols = img.shape
box_filter = np.zeros_like(img, dtype=float)
for row in range(rows):
    for col in range(cols):
        replace = np.round(np.sum(pad_img[row:row+filter_size, col:col+filter_size]))/(filter_size*filter_size)
        box_filter[row][col] = replace
```

```python
plt.imshow(box_filter, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# 2. Weighted average filter
img = cv.imread("/Users/ved/dip/images/img1-1.bmp", 0)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
filter_size = 3
pad_img = np.pad(img, pad_width=1, mode="constant", constant_values=0)
```

```python
weighted_filter = np.array([
    [1,2,1],
    [2,4,2],
    [1,2,1]
])
kernel_value = 16

rows, cols = img.shape
weight_filter_image = np.zeros_like(img, dtype=float)
for row in range(rows):
    for col in range(cols):
        replace = np.round(np.sum(pad_img[row:row+filter_size, col:col+filter_size] *weighted_filter ))/(kernel_value)
        weight_filter_image[row][col] = replace
```

```python
plt.imshow(weight_filter_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# 3. Median Filter
rows, cols = img.shape
median_filter_image = np.zeros_like(img, dtype=float)
for row in range(rows):
    for col in range(cols):
        replace = np.median(pad_img[row:row+filter_size, col:col+filter_size])
        median_filter_image[row][col] = replace
```

```python
plt.imshow(median_filter_image, cmap="gray")
plt.axis("off")
plt.show()
```
