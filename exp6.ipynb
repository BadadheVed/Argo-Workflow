```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/Users/ved/dip/images/img1-1.bmp", cv2.IMREAD_GRAYSCALE)
img1 = np.pad(img, pad_width=2, mode='constant', constant_values=255)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# 1. Point Detection
T = 70
rows, cols = img.shape
filtered_img = np.zeros_like(img)

point_mask = np.array([
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1]
])

point_img = np.zeros_like(point_mask)
```

```python
for row in range(rows):
    for col in range(cols):
        replace = np.sum(img1[row:row+3, col:col+3] * point_mask)
        if replace > T:
            filtered_img[row][col] = 255
        else:
            filtered_img[row][col] = 0

plt.imshow(filtered_img)
plt.axis("off")
plt.show()
```

```python
# 2. Line detection
T = 50
rows, cols = img.shape

#masks
#1. Horizontal mask
hori_mask = ([
    [-1,-1,-1],
    [2,2,2],
    [-1,-1,-1]
])
#2. ratate the mask by 45 degree
dia1_mask = ([
    [-1,-1,2],
    [-1,2,-1],
    [2,-1,-1]
])
#3. Vertical Mask
vert_mask = ([
    [-1,2,-1],
    [-1,2,-1],
    [-1,2,-1]
])
#4. ratate the mask by 45 degree
dia2_mask = ([
    [2,-1,-1],
    [-1,2,-1],
    [-1,-1,2]
])

masks = [hori_mask, dia1_mask, vert_mask, dia2_mask]
merged_image = np.zeros_like(img)
output= []

for mask in masks:
    filtered_image = np.zeros_like(img)
    for row in range(rows):
        for col in range(cols):
            replace = np.sum(img1[row:row+3, col:col+3] * mask)
            if replace > T:
                filtered_image[row][col] = 255
                merged_image[row][col] = 255
            else:
                filtered_image[row][col] = 0
    output.append(filtered_image)

plt.imshow(output[0]) # 1. Hori
plt.axis("off")
plt.show()

plt.imshow(output[1]) # 2. Rotate 45
plt.axis("off")
plt.show()

plt.imshow(output[2]) # 3. Vertical
plt.axis("off")
plt.show()

plt.imshow(output[3]) # 4. Rotate 45
plt.axis("off")
plt.show()

plt.imshow(merged_image) # 5. merged image
plt.axis("off")
plt.show()
```

```python
# 3. Global Thresholding
img1 = cv2.imread("/Users/ved/dip/images/img1-2.bmp", 0)
plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
t = np.mean(img1)
rows, cols = img1.shape
delta = 3

while True:
    G1 = []
    G2 = []
    for row in range(rows):
        for col in range(cols):
            if img1[row][col] > t:
                G1.append(img1[row][col])
            else:
                G2.append(img1[row][col])

    m1 = np.mean(G1)
    m2 = np.mean(G2)
    t_new = (m1 + m2)/2
    if abs(t - t_new) < delta :
        break
    t = t_new

threshold_image = np.zeros_like(img1)
for row in range(rows):
    for col in range(cols):
        if img1[row][col] > t:
            threshold_image[row][col] = 255
        else:
            threshold_image[row][col] = 0
```

```python
plt.imshow(threshold_image) # 5. merged image
plt.axis("off")
plt.show()
```
