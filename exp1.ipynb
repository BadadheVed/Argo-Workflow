```python
# 6. Histogram Function
# 6.0.1 Histogram for grayscale Image
img = cv2.imread("/Users/ved/dip/images/img1-1.bmp")
plt.imshow(img)
plt.axis("off")
plt.show()

histGray = cv2.calcHist([img], [0], None, [256], [0,256])
plt.plot(histGray)
plt.show()
```

```python
# 6.0.2 Histogram of the RGB image
img = cv2.imread("/Users/ved/dip/images/img1-2.bmp")
plt.imshow(img)
plt.axis("off")
plt.show()

colors = ('b', 'g', 'r')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])

plt.title("RGB Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# 6.0.3 Histogram for grayscale Image with mask
img = cv2.imread("/Users/ved/dip/images/img1-1.bmp")
plt.imshow(img)
plt.axis("off")
plt.show()

mask = np.zeros(img.shape[:2], np.uint8)
mask[50:300, 25:350] = 255
mask_image = cv2.bitwise_and(img, img, mask=mask)
plt.imshow(mask_image)
plt.axis("off")
plt.show()

histGray = cv2.calcHist([img], [0], mask, [256], [0,256])
plt.plot(histGray)
plt.show()
```

```python
# 7. Color Conversion
# 7.0.1 Types of conversion
# 1. cv2.COLOR_BGR2GRAY  BGR -> Grayscale
# 2. cv2.COLOR_BGR2RGB   BGR -> RGB
# 3. cv2.COLOR_BGR2HSV   BGR -> HSV
# 4. cv2.COLOR_BGR2LAB   BGR -> LAB color space
# 5. cv2.COLOR_BGR2YCrCb BGR -> YCrCb (used in compression & skin detection)
img = cv2.imread("/Users/ved/dip/images/img1-2.bmp")
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# 7.0.2 BGR to Grayscale
imgBgrToGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imgBgrToGray)
plt.axis("off")
plt.show()
```

```python
# 7.0.3 BGR to RGB
imgBgrToRgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgBgrToRgb)
plt.axis("off")
plt.show()
```

```python
# 8. Thresholding
# 8.0.1 Types of Thresholding
# 1. cv2.THRESH_BINARY     : Above threshold -> 255, below -> 0.
# 2. cv2.THRESH_BINARY_INV : Above threshold -> 0, below -> 255 (inverse of binary).
# 3. cv2.THRESH_TRUNC      : Above threshold -> set to threshold, below -> unchanged.
# 4. cv2.THRESH_TOZERO     : Below threshold -> 0, above -> unchanged.
# 5. cv2.THRESH_TOZERO_INV : Above threshold -> 0, below -> unchanged (inverse of TOZERO).

# 8.0.2 Binary Technique
img = cv2.imread("/Users/ved/dip/images/img1-1.bmp")
plt.imshow(img)
plt.axis("off")
plt.show()

ret, imgThBinary = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
plt.imshow(imgThBinary)
plt.axis("off")
plt.show()
```

```python
# 9. Adding the noise
mean = 0
stddev = 180
img = cv2.resize(img, (200, 200))
noise = np.zeros(img.shape, np.uint8)
noise = cv2.randn(noise, mean, stddev)
noise_img = cv2.add(img, noise)
plt.imshow(noise_img)
plt.axis("off")
plt.show()
```
