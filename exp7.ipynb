```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread("/Users/ved/dip/images/img1-1.bmp", 0)
img = np.pad(img1, pad_width=2, mode='constant', constant_values=0)
plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
# 1. Roberts Cross Operator
rows, cols = img.shape
dgx = np.array([
    [1,0],
    [0,-1]
])
dgy = np.array([
    [0,-1],
    [1,0]
])

dgx_image = np.zeros_like(img)
dgy_image = np.zeros_like(img)
roberts_image = np.zeros_like(img)

for row in range(rows-1):
    for col in range(cols-1):
        r1 = np.sum(img[row:row +2, col:col+2] * dgx)
        r2 = np.sum(img[row:row +2, col:col+2] * dgy)
        roberts_image[row][col] = abs(r1) + abs(r2)
        dgx_image[row][col] = abs(r1)
        dgy_image[row][col] = abs(r2)

plt.imshow(dgx_image)
plt.axis("off")
plt.show()
plt.imshow(dgy_image)
plt.axis("off")
plt.show()
plt.imshow(roberts_image)
plt.axis("off")
plt.show()
```

```python
# 2. Prewitt Operator
rows, cols = img.shape
dgx = np.array([
    [-1,-1,-1],
    [0,0,0],
    [1,1,1]
])
dgy = np.array([
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]
])

dgx_image = np.zeros_like(img)
dgy_image = np.zeros_like(img)
perwitts_image = np.zeros_like(img)

for row in range(rows-2):
    for col in range(cols-2):
        r1 = np.sum(img[row:row +3, col:col+3] * dgx)
        r2 = np.sum(img[row:row +3, col:col+3] * dgy)
        perwitts_image[row][col] = abs(r1) + abs(r2)
        dgx_image[row][col] = abs(r1)
        dgy_image[row][col] = abs(r2)

plt.imshow(dgx_image)
plt.axis("off")
plt.show()
plt.imshow(dgy_image)
plt.axis("off")
plt.show()
plt.imshow(perwitts_image)
plt.axis("off")
plt.show()
```

```python
# 3. Sobel Operator
rows, cols = img.shape
dgx = np.array([
    [-1,-2,-1],
    [0,0,0],
    [1,2,1]
])
dgy = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])

dgx_image = np.zeros_like(img)
dgy_image = np.zeros_like(img)
sobel_image = np.zeros_like(img)

for row in range(rows-2):
    for col in range(cols-2):
        r1 = np.sum(img[row:row +3, col:col+3] * dgx)
        r2 = np.sum(img[row:row +3, col:col+3] * dgy)
        sobel_image[row][col] = abs(r1) + abs(r2)
        dgx_image[row][col] = abs(r1)
        dgy_image[row][col] = abs(r2)

plt.imshow(dgx_image)
plt.axis("off")
plt.show()
plt.imshow(dgy_image)
plt.axis("off")
plt.show()
plt.imshow(sobel_image)
plt.axis("off")
plt.show()
```
